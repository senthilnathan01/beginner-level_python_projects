```python
plt.figure(figsize =(10,6))
plt.plot(y[:, 0], lw=1.5, label = '1st')
plt.plot(y[:, 1], lw=1.5, label = '2nd')
plt.plot(y, 'ro')
plt.legend(loc=0) #best possible locattion
plt.xlabel("Index")
plt.ylabel("Value")
plt.title("A Simple Plot");
```

Looks like we are not able to read the '2nd' data. 

Rescaling is one method. But lets try some other methods. Like:

1. Using two y axes
2. Using 2 subplots(upper/lower, left/right)

```python
fig, ax1 = plt.subplots() 
plt.plot(y[:, 0], 'b', lw =1.5, label = '1st')
plt.plot(y[:, 0], 'ro')
plt.legend(loc=8)
plt.xlabel('index')
plt.ylabel('value 1st')
plt.title('A Simple Plot')
ax2 = ax1.twinx() #Creates a second axis object that shares the x-axis
plt.plot(y[:, 1], 'g', lw = 1.5, label = '2nd')
plt.plot(y[:, 1], 'ro')
plt.legend(loc =0)
plt.ylabel('value 2nd');
```

```python
plt.figure(figsize=(10,6))
plt.subplot(211) #num rows, cols, fignum
plt.plot(y[:, 0], lw=1.5, label = '1st')
plt.plot(y[:, 0], 'ro')
plt.legend(loc=0)
plt.ylabel('value')
plt.title('A Simple Plot')
plt.subplot(212)
plt.plot(y[:, 1], 'g', lw=1.5, label='2nd')
plt.plot(y[:, 1], 'ro')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
```

## Scatter Plot

```python
#using the standard plt.plot function

y = np.random.standard_normal((1000, 2))

plt.figure(figsize=(10, 6))
plt.plot(y[:, 0], y[:, 1], 'ro')
plt.xlabel('1st')
plt.ylabel('2nd')
plt.title('Scatter Plot');
```

```python
# using plt.scatter function. There are some benefits to using this. 

plt.figure(figsize=(10, 6))
plt.scatter(y[:, 0], y[:, 1], marker = 'o')
plt.xlabel('1st')
plt.ylabel('2nd')
plt.title('Scatter Plot');
```

```python
# Among other things, the plt.scatter() plotting function allows the addition of a
# third dimension, which can be visualized through different colors and be described
# by the use of a color bar. 

c = np.random.randint(0, 10, len(y)) #the third dataset

plt.figure(figsize=(10, 6))
plt.scatter(y[:, 0], y[:, 1], c = c, cmap = 'coolwarm', marker = 'o')
plt.colorbar()
plt.xlabel('1st')
plt.ylabel('2nd')
plt.title('Scatter Plot');
```

```python
def func(x):
    return 0.5*np.exp(x)+1
a, b = 0.5, 1.5
x = np.linspace(0, 2)
y = func(x)
Ix = np.linspace(a, b)
Iy = func(Ix)
verts = [(a, 0)]+list(zip(Ix, Iy))+[(b, 0)]
```

```python
from matplotlib.patches import Polygon
fig, ax = plt.subplots(figsize=(10, 6))
plt.plot(x, y, 'b', linewidth = 2)
plt.ylim(bottom=0)
poly = Polygon(verts, facecolor = '0.7', edgecolor='0.5')
ax.add_patch(poly)
plt.text(0.5*(a+b), 1, r'$\int_a^b f(x)\mathrm{d}x$', horizontalalignment='center', fontsize=20)
plt.figtext(0.9, 0.075, '$x$')
plt.figtext(0.075, 0.9, '$f(x)$')
ax.set_xticks((a, b))
ax.set_xticklabels(('$a$', '$b$'))
ax.set_yticks([func(a), func(b)])
ax.set_yticklabels(('$f(a)$', '$f(b)$'));
```

## Static 3D plotting

```python
# Stike values between 50 and 150
# Times to maturity between 0.5 and 2.5 years

strike = np.linspace(50, 150, 24)
ttm = np.linspace(0.5, 2.5, 24)
strike, ttm = np.meshgrid(strike, ttm)

strike[:2].round(1)
```

```python
iv = (strike-100)**2 / (100*strike) / ttm

iv[:5, :3]
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(projection='3d') #sets up canvas for 3D plotting
surf = ax.plot_surface(strike, ttm, iv, rstride=2, cstride=2, cmap=plt.cm.coolwarm, linewidth=0.5, antialiased = True)
ax.set_xlabel('strike')
ax.set_ylabel('time-to-matruity')
ax.set_zlabel('implied volatility')
fig.colorbar(surf, shrink = 0.5, aspect = 5);
```

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.view_init(30, 60) #sets the viewing angle
ax.scatter(strike, ttm, iv, zdir='z', s=25, c='b', marker = '^')
ax.set_xlabel('strike')
ax.set_ylabel('time-to-maturity')
ax.set_zlabel('implied volatility');
```

```python
import pandas as pd
```

```python
import cufflinks as cf
```

```python
import plotly.offline as plyo #imports the offline plotting capabilities as plotly
```

```python
plyo.init_notebook_mode(connected= True) #turns on the notebook plotting mode
```

```python
a = np.random.standard_normal((250, 5)).cumsum(axis=0) 
index = pd.date_range('2019-1-1', #start date for DateTimeIndex object
                      freq = 'B', #business daily frequency
                      periods= len(a)) #number of periods needed
df = pd.DataFrame(100+5*a, #linear transformation of the raw data
                  columns=list('abcde'), #column headers as single character
                  index = index) #first five rows of data
```

```python
df.head()
```

```python
pip install nbformat
```

```python
plyo.iplot(
    df.iplot(asFigure=True), 
    image = 'png', 
    filename = 'ply_01'
)
```

```python
plyo.iplot(df[['a', 'b']].iplot(asFigure = True,
                                theme = 'polar',
                                title = 'A Time Series Plot',
                                xTitle = 'date',
                                yTitle = 'value',
                                mode = {'a':'markers', 'b':'lines+markers'},
                                symbol = {'a':'circle', 'b': 'diamond'},
                                size = 3.5,
                                colors = {'a': 'blue', 'b': 'magenta'}),
                                image = 'png',
                                filename = 'ply_02')
```

```python
plyo.iplot(
    df.iplot(kind='hist', #plotting type
             subplots = True, 
             bins = 15, #number of bars
             asFigure = True),
             image = 'png',
             filename = 'ply_03'
)
```

# Financial Plots

```python
raw = pd.read_csv('fxcm_eur_usd_eod_data.csv', index_col =0, parse_dates =True)
raw.info()
```

```python
quotes = raw[['OpenAsk', 'HighAsk', 'LowAsk', 'CloseAsk']]
quotes = quotes.iloc[-60:] #only 60 are selected for visualization
quotes.tail()
```

```python
qf = cf.QuantFig(
    quotes, # DataFrame is passed to the QuantFig constructor
    title = 'EUR/USD Exchange Rate',
    legend = 'top',
    name = 'EUR/USD'
)
```

```python
plyo.iplot(
    qf.iplot(asFigure = True),
    image = 'png',
    filename = 'qf_01'
)
```

```python
qf.add_bollinger_bands(periods = 10,  #number of periods for the bollinger band
                       boll_std = 2) #number of standard deviations to be used for the band width
```

```python
plyo.iplot(qf.iplot(asFigure = True),
           image = 'png',
           filename = 'qf_02')
```

```python
# RSI

qf.add_rsi(periods = 14, 
           showbands = False)

plyo.iplot(qf.iplot(asFigure = True), image='png', filename = 'qf_03')
```
